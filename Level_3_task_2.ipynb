Restaurant with highest and lowest number of votes

```python
import pandas as pd

# Loading the dataset
df = pd.read_csv('Dataset .csv')

# Setting column names correctly if they differ
votes_col = 'Votes'
name_col = 'Restaurant Name'
rating_col = 'Aggregate rating'

# Cleaning and converting data types
df = df.dropna(subset=[votes_col, name_col, rating_col])
df[votes_col] = pd.to_numeric(df[votes_col], errors='coerce')

# Dropping rows with missing or non-numeric votes
df = df.dropna(subset=[votes_col])

# Top 10 restaurants with highest votes
top_votes = df.sort_values(by=votes_col, ascending=False).head(10)

# Bottom 10 restaurants with non-zero votes
bottom_votes = df[df[votes_col] > 0].sort_values(by=votes_col).head(10)

# Displaying the results
print("🔝 Top 10 Restaurants by Votes:")
print(top_votes[[name_col, votes_col, rating_col]])

print("\n🔻 Bottom 10 Restaurants by Votes (non-zero):")
print(bottom_votes[[name_col, votes_col, rating_col]])
```

Correlation between number of votes and ratings

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_csv('Dataset .csv')

#  Set column names
votes_col = 'Votes'
rating_col = 'Aggregate rating'

# Cleaning and converting columns
df = df.dropna(subset=[votes_col, rating_col])
df[votes_col] = pd.to_numeric(df[votes_col], errors='coerce')
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')
df = df.dropna(subset=[votes_col, rating_col])

# Scatter plot: Votes vs Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=votes_col, y=rating_col, alpha=0.5)
plt.title("Votes vs Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.grid(True)
plt.show()

#  Correlation
correlation = df[votes_col].corr(df[rating_col])
print(f"\n📊 Correlation between Votes and Rating: {correlation:.2f}")
```
