IDENTIFY RESTAURANT CHAINS

```python
# Step 1: Identifying Restaurant Chains
import pandas as pd

# Loading the dataset
df = pd.read_csv('Dataset .csv')

# Counting how many times each restaurant name appears
name_counts = df['Restaurant Name'].value_counts()

# Identifying potential chains (appearing more than once)
chains = name_counts[name_counts > 1]

print("Potential restaurant chains:")
print(chains)

# Optionally: View as a DataFrame
chains_df = chains.reset_index()
chains_df.columns = ['Restaurant Name', 'Number of Locations']
chains_df.head()
```

```python
# @title Number of Locations

from matplotlib import pyplot as plt
chains_df['Number of Locations'].plot(kind='line', figsize=(8, 4), title='Number of Locations')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Step 2: Analyzing the Ratings and Popularity of Chains

# Calculating average rating and total votes for each restaurant
agg_metrics = df.groupby('Restaurant Name').agg(
    Number_of_Locations=('Restaurant Name', 'count'),
    Average_Rating=('Aggregate rating', 'mean'),
    Total_Votes=('Votes', 'sum')
).reset_index()

# Filtering only chains (more than one location)
chain_analysis = agg_metrics[agg_metrics['Number_of_Locations'] > 1]

# Sort by popularity (votes)
chain_analysis_sorted = chain_analysis.sort_values(by='Total_Votes', ascending=False)

# Show top 10 chains
chain_analysis_sorted.head(10)
```

VISUALIZE TOP 10 CHAINS

```python
# Plot top 10 chains by average rating
import matplotlib.pyplot as plt

top_chains = chain_analysis_sorted.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top_chains['Restaurant Name'], top_chains['Average_Rating'], color='skyblue')
plt.xlabel('Average Rating')
plt.title('Top 10 Restaurant Chains by Average Rating')
plt.gca().invert_yaxis()
plt.show()
```
