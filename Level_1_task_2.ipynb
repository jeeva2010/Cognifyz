TOP 10 CITIES WITH NUMBER OF RESTAURANTS


```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading dataset
df = pd.read_csv('Dataset .csv')

# Counting restaurants per city
city_counts = df['City'].value_counts()

# Getting city with highest number of restaurants
top_city_by_count = city_counts.idxmax()
top_count = city_counts.max()

# Displaying the result
print(f"City with the highest number of restaurants: {top_city_by_count} ({top_count} restaurants)")

# Bar chart: Top 10 cities by number of restaurants
plt.figure(figsize=(10, 6))
city_counts.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

AVERAGE of 10 cities


```python
# Clean rating data: remove non-numeric/missing ratings
df_cleaned = df[pd.to_numeric(df['Aggregate rating'], errors='coerce').notnull()]
df_cleaned['Aggregate rating'] = df_cleaned['Aggregate rating'].astype(float)

# Grouping by city and calculate the average rating
avg_rating_by_city = df_cleaned.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

# Display average ratings for first 10 cities
print("Average Rating per City (Top 10):")
print(avg_rating_by_city.head(10))
```

CITY WITH HIGHEST AVERAGE RATING

```python
# Get city with the highest average rating
top_city_by_rating = avg_rating_by_city.idxmax()
top_avg_rating = avg_rating_by_city.max()

# Displaying the result
print(f"City with the highest average rating: {top_city_by_rating} ({top_avg_rating:.2f})")

# Bar chart: Average rating for top 10 most frequent cities
top_10_cities_by_count = city_counts.head(10).index
avg_ratings_top_cities = avg_rating_by_city[top_10_cities_by_count]

plt.figure(figsize=(10, 6))
avg_ratings_top_cities.plot(kind='bar', color='green')
plt.title('Average Rating in Top 10 Cities by Restaurant Count')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
